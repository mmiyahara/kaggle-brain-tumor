```python
b, g, r = cv.split(img)
# or b = img[:, :, 0]
print(b.shape)
print(g.shape)
print(r.shape)
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(2, 3, 1)
ax1.axis('off')
ax1.imshow(b)
ax2 = fig.add_subplot(2, 3, 2)
ax2.axis('off')
ax2.imshow(g)
ax3 = fig.add_subplot(2, 3, 3)
ax3.axis('off')
ax3.imshow(r)
ax4 = fig.add_subplot(2, 3, 5)
ax4.axis('off')
ax4.imshow(img)
```
