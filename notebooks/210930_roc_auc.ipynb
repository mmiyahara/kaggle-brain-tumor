```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
answers = np.array([0, 0, 0, 0, 0, 1, 1, 0, 1, 1])

# 悪い判定
bad_preds = np.array([0.8, 0.7, 0.9, 0.1, 0.2, 0.9, 0.7, 0.2, 0.3, 0.1])
# 良い判定
good_preds = np.array([0.1, 0.2, 0.1, 0.6, 0.4, 0.9, 0.7, 0.2, 0.8, 0.3])

b_fpr, b_tpr, _ = roc_curve(answers, bad_preds)
g_fpr, g_tpr, _ = roc_curve(answers, good_preds)
b_auc = auc(b_fpr, b_tpr)
g_auc = auc(g_fpr, g_tpr)
plt.plot(b_fpr, b_tpr, label=f'ROC curve (area = {b_auc:.2f})')
plt.plot(g_fpr, g_tpr, label=f'ROC curve (area = {g_auc:.2f})')
plt.plot([0, 1], [0, 1])
plt.legend()
```
