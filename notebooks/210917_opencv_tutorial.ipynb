```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
x = np.uint8([250])
y = np.uint8([10])
z = np.uint8([5])

# NumPy の足し算では 255 を超えたとき、飽和した分が結果に
print(x+y)
print(x+z)

# cv2 の足し算では 255 を超えたとき、255 に
print(cv.add(x, y))
print(cv.add(x, z))
```

```python
org = cv.imread("lenna.png")
org.shape
```

```python
print(org.min(), org.max())
```

```python
rand_img = np.random.randint(0, 512, size=(512, 512, 3))
mixed = cv.addWeighted(org, 0.7, rand_img, 0.3, 0, dtype=0)
fig = plt.figure(figsize=(16, 12))
for i, image in enumerate((org, rand_img, mixed)):
    fig.add_subplot(1, 3, i+1)
    plt.imshow(image)
    plt.axis('off')
fig.tight_layout()
plt.show()
```
